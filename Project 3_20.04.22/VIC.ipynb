```python
#Dropping Na values - how = all means drop only when all values are na - which is rare because the first five columns are populated 
drop_all_na = VIC.dropna(how = 'all')
drop_all_na
```

```python
#Dropping Na values - how = any means drop when even a single value is na - this doesn't return anything because there is a column called Ã¡uction number withdrawn'which is empty
drop_all_na = VIC.dropna(how = 'any')
drop_all_na
```

```python
# drop na's based on a subset
VIC_01 = VIC.dropna(subset=['highestSoldPrice', 'numberSaleListing', 'lowestSaleListingPrice'], how= 'all') 
VIC_01
```

```python
#We have 1243 values left from 2300 values
VIC_01.to_csv('VIC_01.csv',index=False)
VIC_01
```

```python
#Created connection with local database SQL
rds_connection_string = "postgres:password@localhost:5432/Victoria_db"
engine = create_engine(f'postgresql://{rds_connection_string}')
```

```python
engine.table_names()
```

```python
#Use pandas to load csv converted dataframe to database
VIC_01.to_sql(name='Victoria', con=engine, if_exists='append', index=False)
```

```python
#Confirm data has been added by querying the table
pd.read_sql_query('select * from Victoria', con=engine).head()
```
